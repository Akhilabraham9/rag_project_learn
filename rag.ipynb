# RAG - Retrieval Augumented Generation

* It is process of optimizing the output of a Large Language Model, so it references an authoritative knowledge base outside of its training data sources before generating a response. 
* LLM are trained on vast volumes of data and use billions of parameters to generate original output for tasks like answering question, translating languages and completing sentences. 
* RAG extends the already powerful capabilities of LLMs to specific domains or an organization's internal knowledge base, all without the need to retrain the model. It is a cost effective approach to improving LLM output so it remains relevant, accurate, and useful in various contexts.
